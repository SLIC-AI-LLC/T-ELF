# Libraries

```python
import os
import pandas as pd
```

## 1. Load Data

```python
df = pd.read_csv(os.path.join("..", "..", "data", "sample2.csv"))
df.info()
```

## 2. Create SLIC Authors Map

```python
from TELF.pre_processing import Orca
orca = Orca(verbose=True)
```

```python
slic_df = orca.run(df)
slic_df.info()
```

```python
slic_df
```

### Sort by most Affiliations

```python
slic_df = slic_df.sort_values(by='scopus_affiliations', key=lambda col: col.map(lambda x: len(x) if pd.notna(x) else 0), ascending=False).reset_index(drop=True)
slic_df
```

```python
slic_df.iloc[0].scopus_affiliations
```

### Look at Cases where S2 IDs are Duplicated

```python
slic_df.loc[slic_df.s2_ids.str.contains(';') & pd.notna(slic_df.s2_ids)]
```

### Look at Cases where Scopus IDs are Duplicated

```python
slic_df.loc[slic_df.scopus_ids.str.contains(';') & pd.notna(slic_df.scopus_ids)]
```

## 3. Propogate SLIC Mapping to DataFrame

```python
df = orca.apply(df)
df.info()
```
